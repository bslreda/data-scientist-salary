### handling tthe missing values and features engeneering

```python
# date we should make the date format
df['date'].describe()
```

```python
def dt(df):
    df['Date']=df['date'].apply(lambda x: x.split('T')[0])
    return df
```

```python
dt(df)
```

```python
df['year']=df['Date'].apply(lambda x: x[0:4])
```

```python
df['month']=df['Date'].apply(lambda x: x[4:6])
```

```python
df['day']=df['Date'].apply(lambda x: x[6:])
```

```python
df['myDt']=pd.to_datetime(df[['year','month','day']])
```

```python
df['myDt']
```

```python
# drop the old date
df.drop('date',axis='columns', inplace=True)
```

```python
# drop id  we dont want to use it
df.drop('id',axis='columns', inplace=True)
```

```python
# drop zip code  we dont want to use it
df.drop('zipcode',axis='columns', inplace=True)
```
